## 3A - Data Wrangling with Python: Activity 1 page 17

```python
#Step 1 - Create a list of 100 random numbers
random_list =  [random.randint(0,101) for x in range (0,101)]
random_list
```

```python
#Step 2 - Create a new list from this random list, with numbers that are divisible by 3
div3_list = [i for i in random_list if i % 3 ==0]
div3_list
```

```python
#Step 3 - Calculate the length of these two lists and store the difference in a new variable
len_1 = len(random_list)
len_2 = len(div3_list)
diff_len = len_1 - len_2
diff_len
```

```python
#Step 4 - Using a loop, perform steps 2 and 3 and find the difference variable three times
loop_3 = 3 #defines how many times I want loop to run
diff_list = []
for i in range(0, loop_3):
    random_list2 = [random.randint(0,101) for x in range (0,101)]
    div_list3 = [i for i in random_list2 if i % 3 == 0]
    
    len1_list = len(random_list2)
    len2_list = len(div_list3)
    difference = len1_list - len2_list
    
    diff_list.append(difference)
diff_list
```

```python
#Step 5 - Find the arithmetic mean of these three difference values
mean_diff = sum(diff_list) / float(len(diff_list)) #Got some help from solution code in the book on pg 342
mean_diff
```
