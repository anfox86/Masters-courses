# Data Wrangling with Python: Activity 6, page 171

```python
#1 - Already loaded all the libraries I think I'll need for this exercise
#2 - Read the adult income dataset from the URL
df_adult = pd.read_csv('adult_income_data.csv')
df_adult.head()
#Tried using url and then calling the url with sep = "," but I kept getting a parsing error so did it this way
```

```python
#3 - Create a script that will read a text file line by line
col_names = [] #Create variable to read column names
with open('adult_income_names.txt','r') as f:
    for line in f:
        f.readline() #Reads line by line
        split = line.split(":")[0]
        col_names.append(split) #Adds each line to the existing variable
col_names
```

```python
#4 - Add a name of Income for the response variable to the dataset
col_names.append('Income')
col_names
```

```python
#5 - Find the missing values
df_adult.isnull() #Realized I was still getting the old column names that don't explain anything
df_income = pd.read_csv('adult_income_data.csv', names = col_names)
df_income.head()
```

```python
#Tried after fixing dataframe
df_income.isnull() #This showed me the head and tail with all falses as far as I could see. Checked solution code to see how they did it and updated 
df_income.isnull().sum() #Adding the sum() showed nulls by column which is much easier to digest.
```

```python
#6 - Create a DataFrame with only age, education, and occupation by using subsetting
df_sub = df_income[['age', 'education', 'occupation']]
df_sub.head()
```

```python
#7 - Plot a histogram of age with a bin size of 20
plt.hist(df_sub.age, bins = 20)
plt.show()
```

```python
#8 - Create a function to strip the whitespace characters
def strip_whitespace(s):
    return s.strip()
#9 - Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column, and drop the new column
#Definitely used solution code here as I couldn't wrap my mind on the initial ask. Once I saw it and spoke it outloud it made sense
#education
df_sub['education_stripped']=df_income['education'].apply(strip_whitespace) #creating new column and removing whitespace
df_sub['education']=df_sub['education_stripped'] #updating existing column with the new 
df_sub.drop(labels=['education_stripped'],axis=1,inplace=True) #Removing new column name as old one was updated

#occupation
df_sub['occupation_stripped']=df_income['occupation'].apply(strip_whitespace) #creating new column and removing whitespace
df_sub['occupation']=df_sub['occupation_stripped'] #updating existing column with the new
df_sub.drop(labels=['occupation_stripped'], axis=1, inplace=True) #Removing new column name as old one was updated
```

```python
#10 - Find the number of people who are aged between 30 and 50
df_age = df_sub[(df_sub.age>=30)&(df_sub.age<=50)] #Created new variable for specified ages using the subset
df_age
```

```python
#11 - Group the records based on age and education to find how the mean age is distributed
df_sub.groupby(['education']).mean()
```

```python
#Step 12 - Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on average and which profession has its largest share of the workforce above the 75th percentile.
df_sub.groupby('occupation').describe()['age']
```

```python
#Step 13 - Use subset and groupby to find outliers
#This is kind of answered in the groupby occupation above. I have a hard time believing that armed forces only has 9 people. 

#Step 14 - Plot the values on a bar chart
occupation_age = df_sub.groupby('occupation').describe()['age'] #Needed to create a variable for this so it would be easier to call into the bar chart

plt.barh(y = occupation_age.index,width = occupation_age['count']) #Had issues with this part. Had to google the right plot function. Then didn't realize I had to do it by y axis not x. Ended up checking solution code to figure out how they did it
plt.yticks(fontsize=10)
plt.show()
```

# Create a series and practice basic arithmetic steps

```python
#Series 1
series_1 = pd.Series([7.3, -2.5, 3.4, 1.5], index = ['a', 'c', 'd', 'e'])
series_1
```

```python
#Series 2
series_2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index = ['a', 'c', 'e', 'f', 'g'])
series_2
```

```python
#Add Series 1 and Series 2 together and print the results
add = series_1.add(series_2, fill_value = 0) 
print(add)
#The first time I did this I got a bunch of NaN values. Went to google and found https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.add.html
```

```python
#Subtract Series 1 from Series 2 and print the results
#Used same idea only with sub from above. Also found the function on pandas
subtract = series_1.sub(series_2, fill_value = 0)
print(subtract)
```
