```python
#Decided to use So Much Candy. Opening xlsx file from website
url = 'https://www.scq.ubc.ca/wp-content/uploads/2017/10/candyhierarchy2017.xlsx'
df = pd.read_excel(url, header=0)
df #I show 2460 rows and 120 columns
```

## Chapter 7

```python
#Filter out missing data
#This will show me how many null values are in each column
for c in df.columns:
    miss = df[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has no missing values".format(c))
```

```python
#To remove them I'm using dropna but specifying a threshold. Using a higher number because some of the columns barely have any data
df_drop = df.dropna(thresh=100)
df_drop #Now I have 1537 rows and 120 columns. When I tried just dropna without specifying a threshold it went to no data. I'm assuming at this point every row has some piece of missing data which is why I moved to this method.
```

```python
#Drop duplicates
#Since many of the other columns are likely to have duplicate values I only want to drop any duplicate Internal ID as that is the primary key for this data
df_candy = df_drop.drop_duplicates(['Internal ID'])
df_candy #Had no duplicates among Internal ID
```

## Chapter 8

```python
#First I'm going to split the original dataframe into two separate subsets so I can practice putting them back together again
df_male = df_candy[df_candy["Q2: GENDER"].isin(["Male"])] #Created a subset of only male responses
df_female = df_candy[df_candy["Q2: GENDER"].isin(["Female"])] #Created a subset of only female responses
df_female.head() #Just taking a peek to make sure it worked as expected
```

```python
#Now to merge these I'm going to a many-to-one join using gender
df_combo = df_female.combine_first(df_male)
df_combo
#Struggled with this one at first. Tried using merge, but ended up with 239 columns and no rows. Looked further into the chapter and saw data that has overlap use combine_first pg 245
```

```python
#Reshaping the data using hierarchical indexing pg 247
result = df_combo.stack()
result #Now data is rotated 
```

```python
#Reshaping using unstack
reshaped = result.unstack()
reshaped #Data is pivoted back although not sure why Internal ID is showing like it is now
```
